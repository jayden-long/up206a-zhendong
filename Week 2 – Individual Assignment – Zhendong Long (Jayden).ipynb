## Operation: Triming the data set

```python
# I want to trim the original data set before the next steps:

# I want to focus on the name

parkandride_trimmed = parkandride.drop(['source', 'ext_id', 'cat1','cat2','cat3','addrln1','addrln2','state','hours','phones', 'email', 'url', 'info1','info2', 'post_id','descriptio','zip','link', 'use_type','date_updat','dis_status','POINT_X','POINT_Y', 'geometry'], axis=1)

parkandride_trimmed
```

```python
parkandride_trimmed.columns = ['No','Owner','Site','City','Latitude','Longitude']

parkandride_trimmed
```

**Finding:** For me, the name of each location, the owner, the geo-location are the most important information for me and the other information are eliminated because they are not useful in terms of my project goal.

## Operation: Querying

```python
# I want to know the list of park and ride sites within the city of "Los Angeles"

parkandride_trimmed.query("City == 'Los Angeles'")
```

**Finding:** The city of LA has the most park and ride locations in the county. By doing the query, I am able to see the entire list of LA city's locations. 

## Operation: Visualize all park and ride sites in LA county

```python
# First, I want to visualize the data to get a sense what it may look like.

parkandride.plot()
```

```python
# But, I want to viualize all sites on the base map - a real map:

# first, I need to load the module:

import folium

# then, default folium map:

m = folium.Map()

# then, I want to see it:

m
```

```python
# I want to find the center point of my sites:

# first, to find the average latitude: 

latitude = parkandride_trimmed.Latitude.mean()

# then, to find the average longitutde: 

longitude = parkandride_trimmed.Longitude.mean()
```

```python
latitude
```

```python
longitude
```

**Note:** This is the geo-location I want to put in the center of the map.

```python
# Then, I want to zoom in the map at my center point:

m = folium.Map(location=[34.114218906102934, -118.21010969066172], zoom_start=10)

m
```

```python
# I want to make markers for all sites:

# I have to create a for loop:

for index, row in parkandride_trimmed.iterrows():
    print(row.No, row.Site, row.City, row.Latitude, row.Longitude)
    
```

**Note:** There are over 100 locations and I want the machine to repeat the same function/code for every single location. Therefore, I have to create a for loop code to tell the machine to read so.

```python
# Now, I want to bring those for loop into the map by telling the machine to mark every single value in the loop.

for index, row in parkandride_trimmed.iterrows():
    folium.Marker([row.Latitude, row.Longitude], popup=row.Site).add_to(m)

m
```

**Final Product:** This is the final result of data analysis and visulization. All park and ride locations in the county are clearly shown in the real world map.
